## Next Section is Data Aggregation

We have 36717 cars of 40 brands left so far. I decided to choose top 20 brands of them.

```python
autos['brand'].describe()
```

```python
data_aggregation = {}
```

```python
autos['brand'].value_counts()
```

```python
brands = autos['brand'].value_counts().index
new_brands = brands[:20]
```

```python
for brand in new_brands:
    series = autos[autos['brand'] == brand]
    mean_series = series['price'].mean()
    data_aggregation[brand] = mean_series
```

```python
k = 0
new_data = {}
for key in data_aggregation:
    if key in new_brands:
        new_data[key] = data_aggregation[key]
new_data
```

```python
sorted_data = sorted(new_data.items(), key=lambda val: val[1])
sorted_data.reverse()
mean_prices = {}
for data in sorted_data:
    mean_prices[data[0]] = data[1]
mean_prices
```

## Conclusion
From the _sorted_data_ dictionary, it can be seen that the mean price of the top-20 put on is between 3500 and 12200 euros.
Sonstige_autos and Mini are expensive, because they are rare cars. Surprisingly, top-20 french cars are cheaper than others.
Audi, BMW, and Mercedes are expensive, while Ford and Opel are much more cheaper. Most abundant brand - Volskwagen is in between.

```python
bmp_series = pd.Series(mean_prices)
print(bmp_series)
```

```python
odom_means = {}
for brand in new_brands:
    odom = autos[autos['brand'] == brand]
    mean_odom = odom['odometer_km'].mean()
    odom_means[brand] = mean_odom
odom_means
```

```python
bmo_series = pd.Series(odom_means)
bmo_series
```

```python
bmp_df = pd.DataFrame(bmp_series, columns=['mean_price'])
```
